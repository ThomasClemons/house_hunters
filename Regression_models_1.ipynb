```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
# Set Max display rows and columns
pd.options.display.max_rows=None
pd.options.display.max_columns=None
```

```python
# Import the data
X_train = pd.read_csv('Resources/housing_df_train_clean_revised.csv')
X_test = pd.read_csv('Resources/housing_df_test_clean_revised.csv')
y_test = pd.read_csv('Resources/housing_df_y_test.csv')
y_train = pd.read_csv('Resources/housing_df_y_train.csv')
print("Input file has been read in")
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import cross_val_score
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
from xgboost import XGBRegressor
```

```python
# Create and train the Random Forest Regression model
rf = RandomForestRegressor(n_estimators=100, random_state=42)  
rf.fit(X_train, y_train)

# Make predictions
y_pred = rf.predict(X_test)

# Evaluate the model
r_squared = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

# Display the results
print(f"R-squared: {r_squared}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
# Plotting actual vs. predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='brown')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted (Random Forest Regression)')
plt.show()
```

```python
lasso = Lasso(alpha=1.0)  
lasso.fit(X_train, y_train)

# Make predictions
y_pred = lasso.predict(X_test)

# Evaluate the model
r_squared = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)

# Display the results
print(f"R-squared: {r_squared}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
# Plotting actual vs. predicted
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, color='blue')
plt.xlabel('Actual')
plt.ylabel('Predicted')
plt.title('Actual vs. Predicted (Lasso Regression)')
plt.show()
```

Using Ridge, XGBOOST, LIGHTGBM, and CatBOOST Regression Models 

```python
# Create and train the Ridge regression model
ridge = Ridge(alpha=1.0)  
ridge.fit(X_train, y_train)

# Make predictions
y_pred = ridge.predict(X_test)

# Evaluate the model
r_squared = r2_score(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = sqrt(mse)

# Display the results
print(f"R-squared: {r_squared}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5)
plt.title('Actual vs Predicted Sale Prices')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.show()
```

```python
residuals = y_test - y_pred
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals, alpha=0.5)
plt.title('Residual Plot')
plt.xlabel('Predicted Sale Price')
plt.ylabel('Residuals')
plt.axhline(y=0, color='r', linestyle='--')
plt.show()
```

```python
# X_train and y_train 
ridge = Ridge(alpha=1.0)  
ridge.fit(X_train, y_train)

# DataFrame to store the coefficients
coefficients = pd.DataFrame(ridge.coef_.reshape(-1, 1), index=X_train.columns, columns=['Coefficient'])

# Plot the coefficients
coefficients.plot(kind='bar', figsize=(14, 6))
plt.title('Feature Coefficients')
plt.xlabel('Feature')
plt.ylabel('Coefficient')
plt.xticks(rotation=45, ha='right')  
plt.tight_layout()  
```

```python
# Cross-validate the model to get an idea of its generalization performance
scores = cross_val_score(ridge, X_train, y_train, cv=5, scoring='neg_mean_squared_error')

# Calculate the mean and standard deviation of the cross-validation scores
mean_score = np.mean(scores)
std_score = np.std(scores)

print(f"Mean Cross-validated MSE: {mean_score}")
print(f"Standard Deviation of Cross-validated MSE: {std_score}")
```

```python
# Create an instance of XGBoost regression
xgb_reg = XGBRegressor()

# Fit the model on the training data
xgb_reg.fit(X_train, y_train)

# Make predictions on the test data
y_pred = xgb_reg.predict(X_test)

# Calculate R-squared
r_squared = r2_score(y_test, y_pred)

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error
rmse = sqrt(mse)

print(f"R-squared: {r_squared}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, c='green')  
plt.title('Actual vs Predicted Sale Prices (XGBoost)')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.show()
```

```python
# Calculate residuals
residuals_xgb = y_test['SalePrice'] - y_pred

# Scatter plot of residuals
plt.figure(figsize=(10, 6))
plt.scatter(y_pred, residuals_xgb, alpha=0.5, c='red')  
plt.title('Residual Plot (XGBoost)')
plt.xlabel('Predicted Sale Price')
plt.ylabel('Residuals')
plt.axhline(y=0, color='black', linestyle='--')  
plt.show()
```

```python
# Perform 5-fold cross-validation
scores = cross_val_score(xgb_reg, X_train, y_train, cv=5, scoring='neg_mean_squared_error')

# Calculate the mean and standard deviation of the cross-validation scores
mean_score = np.mean(scores)
std_score = np.std(scores)

print(f"Mean Cross-validated MSE: {mean_score}")
print(f"Standard Deviation of Cross-validated MSE: {std_score}")
```

```python
# Create an instance of CatBoost regression
catboost_reg = CatBoostRegressor()

# Fit the model on the training data
catboost_reg.fit(X_train, y_train)

# Make predictions on the test data
y_pred = catboost_reg.predict(X_test)

# Calculate R-squared
r_squared = r2_score(y_test, y_pred)

# Calculate Mean Absolute Error
mae = mean_absolute_error(y_test, y_pred)

# Calculate Mean Squared Error
mse = mean_squared_error(y_test, y_pred)

# Calculate Root Mean Squared Error
rmse = sqrt(mse)

print(f"R-squared: {r_squared}")
print(f"Mean Absolute Error: {mae}")
print(f"Mean Squared Error: {mse}")
print(f"Root Mean Squared Error: {rmse}")

```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred, alpha=0.5, c='orange')  # Change 'orange' to any color you prefer
plt.title('Actual vs Predicted Sale Prices (CatBoost)')
plt.xlabel('Actual Sale Price')
plt.ylabel('Predicted Sale Price')
plt.show()
```

```python
# DataFrame to store the feature coefficients
feature_importances_catboost = pd.DataFrame(catboost_reg.feature_importances_, index=X_train.columns, columns=['Importance'])

# Sort the features by importance
feature_importances_catboost = feature_importances_catboost.sort_values('Importance', ascending=False)

# Plot the feature importances
plt.figure(figsize=(14, 6))
plt.bar(feature_importances_catboost.index, feature_importances_catboost['Importance'])
plt.title('Feature Importances (CatBoost)')
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
